# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
directrs_films = []
for index,f in data.iterrows():
    for d in f.director.split('|'):
        directrs_films += [(f.imdb_id, d)]
directors_df = pd.DataFrame(directrs_films, columns=['imdb_id', 'single_director'])
directors_df.describe()
count_y = lambda x: len([ i for i in x if i  > 0])
count_n = lambda x: len([ i for i in x if i <= 0])
data['income_y'] = data['income']
data['income_n'] = data['income']
filter=['Quentin Tarantino', 'Steven Soderbergh', 'Robert Rodriguez', 'Christopher Nolan', 'Clint Eastwood']
res=directors_df.merge(data, on='imdb_id').groupby('single_director').aggregate({"imdb_id": len, "income_y": count_y, "income_n": count_n})
res['p'] = res.income_y / res.imdb_id
res.sort_values(by='p', ascending=False)
res.reset_index(inplace=True)
res[res.single_director.isin(filter)].sort_values(by='p', ascending=False)
answer_ls.append(4)
```
